### 8. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the number of releases in a year as a parameter of popularity of a genre

Hint

Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 2: Use groupby on the relevant column and use sum() on the same to find out the nuumber of releases in a year/genre.  

```python
#Your answer here
dfMovies['release year'] = dfMovies['release date'].apply(lambda x : x.split('-')[2])
dfGenrePopularity = dfMovies.iloc[:,3:].groupby(["release year"]).sum()
print('\033[1mGenre Popularity Over Year:\033[0m')
plt.figure(figsize=(16,10))
sns.heatmap(dfGenrePopularity,cmap='coolwarm')
plt.show()
```

### 9. Find the top 25 movies according to average ratings such that each movie has number of ratings more than 100 - 10 marks

Hint : 
1. First find the movies that have more than 100 ratings(use merge, groupby and count). Extract the movie titles in a list.
2. Find the average rating of all the movies and sort them in the descending order. You will have to use the .merge() function to reach to a data set through which you can get the names and the average rating.
3. Use isin(list obtained from 1) to filter out the movies which have more than 100 ratings.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
#your answer here

# creating dataframe for Movies and Rating by merging them based on Movie Id
# and grouping merged output by movie id and movie name. purposefully including movie id for more accurate result
dfMovieRatingGroup = pd.merge(dfMovies[['movie id','movie title']],
                 dfRatings[['movie id','rating']],
                 on='movie id').groupby(['movie id','movie title'])

# creating dataframe using size for value count and resetting index to count
dfRatingCount = dfMovieRatingGroup.size().reset_index(name='count')

# creating dataframe where rating count is greater than 100
dfCount = dfRatingCount.loc[dfRatingCount['count'] > 100]

# creating dataframe for mean result
dfRatingMean = dfMovieRatingGroup.mean()

# merging above two dataframes on movie id and sort it by descending order of rating and selecting top 25 rows
pd.merge(dfCount, dfRatingMean, on='movie id').sort_values('rating', ascending=False).head(25)
```
